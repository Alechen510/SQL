<span style="font-size: 16.000000pt; font-family: 'Cambria'">32. High-value customers</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">We want to send all of our high-value customers a special VIP gift. We're defining high- value customers as those who've made at least 1 order with a total value (not including the discount) equal to $10,000 or more. We only want to consider orders made in the year 2016.&nbsp;</span>

```sql
SELECT 
    C.CustomerID,
    C.CompanyName, 
    O.OrderID,
    TotalOrderAmount = SUM(OD.Quantity * OD.UnitPrice)
FROM Orders O
JOIN OrderDetails OD ON O.OrderID = OD.OrderID
JOIN Customers C on C.CustomerID = O.CustomerID AND YEAR(O.OrderDate) = 2016
GROUP BY C.CustomerID,C.CompanyName, O.OrderID
HAVING SUM(OD.Quantity * OD.UnitPrice) >= 10000
ORDER BY TotalOrderAmount DESC 
```

<span style="font-size: 16.000000pt; font-family: 'Cambria'">33. High-value customers</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">—</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">total orders</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">The manager has changed his mind. Instead of requiring that customers have at least one individual order totaling $10,000 or more, he wants to define high-value customers differently. Now, high value customers are customers who have orders totaling $15,000 or more in 2016. How would you change the answer to the problem above?&nbsp;</span>

```sql
SELECT 
    C.CustomerID, 
    C.CompanyName, 
    TotalOrderAmount = SUM(OD.Quantity * OD.UnitPrice)
FROM Orders O 
JOIN OrderDetails OD ON O.OrderID = OD.OrderID
JOIN Customers C ON C.CustomerID = O.CustomerID AND YEAR(O.OrderDate) = 2016
GROUP BY C.CustomerID, C.CompanyName
HAVING SUM(OD.Quantity * OD.UnitPrice) >= 15000
ORDER BY TotalOrderAmount DESC
```

<span style="font-size: 16.000000pt; font-family: 'Cambria'">34. High-value customers</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">—</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">with discount</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">Change the answer from the previous problem to use the discount when calculating high- value customers. Order by the total amount, taking into consideration the discount.&nbsp;</span>

```sql
SELECT 
    C.CustomerID, 
    C.CompanyName, 
    TotalWithoutDiscount = SUM(OD.Quantity * OD.UnitPrice), 
    TotalWithDiscount = SUM(OD.Quantity * OD.UnitPrice *(1 - OD.Discount))  ----- Percentage off 
FROM Orders O 
JOIN OrderDetails OD ON O.OrderID = OD.OrderID
JOIN Customers C ON C.CustomerID = O.CustomerID AND YEAR(OrderDate) = 2016
GROUP BY C.CustomerID, C.CompanyName
HAVING SUM(OD.Quantity * OD.UnitPrice *(1 - OD.Discount))  >= 15000 
ORDER BY TotalWithDiscount DESC 
```

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">Andrew Fuller, the VP of sales at Northwind, would like to do a sales campaign for existing customers. A customer with purchase from 1,000 to 5,000 would be in the “Medium” group, and so on.</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">A good starting point for this query is the answer from the problem </span> <span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">“</span><span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">High-value customers</span><span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">—</span><span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">total orders</span><span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">”</span><span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">. <span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">Order the results by CustomerID.&nbsp;</span> 

<span style="font-size: 16.000000pt; font-family: 'Cambria'">49. Customer grouping</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">—</span><span style="font-size: 16.000000pt; font-family: 'Cambria'">fix null&nbsp;</span> 

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">There's a problem with the answer to the previous question. The CustomerGroup value for one of the rows is null.</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">Fix the SQL so that there are no nulls in the CustomerGroup field.&nbsp;</span>

```sql
WITH TotalOrderAmount AS (
SELECT 
    O.CustomerID,
    C.CompanyName,
    TotalOrderAmount = SUM(Quantity*UnitPrice)
FROM OrderDetails OD 
JOIN Orders O ON OD.OrderID = O.OrderID AND YEAR(O.OrderDate) = 2016
JOIN Customers C ON O.CustomerID = C.CustomerID
GROUP BY O.CustomerID, C.CompanyName
)


SELECT 
    *,
   CustomerGroup = CASE 
        WHEN TotalOrderAmount >= 0 AND TotalOrderAmount <= 1000 THEN 'Low'
        WHEN TotalOrderAmount >= 1000 AND TotalOrderAmount <=5000 THEN 'Medium'
        WHEN TotalOrderAmount >=5000 AND TotalOrderAmount <=10000 THEN 'High'
        ELSE 'Very High'
    END
FROM TotalOrderAmount



```

<span style="font-size: 16.000000pt; font-family: 'Cambria'">50. Customer grouping with percentage</span>

<span style="font-size: 12.000000pt; font-family: 'TimesNewRomanPSMT'">Based on the above query, show all the defined CustomerGroups, and the percentage in each. Sort by the total in each group, in descending order.&nbsp;</span>

```sql
WITH TotalOrderAmount AS (
SELECT 
    O.CustomerID,
    C.CompanyName,
    TotalOrderAmount = SUM(Quantity*UnitPrice)
FROM OrderDetails OD 
JOIN Orders O ON OD.OrderID = O.OrderID AND YEAR(O.OrderDate) = 2016
JOIN Customers C ON O.CustomerID = C.CustomerID
GROUP BY O.CustomerID, C.CompanyName
),
CustomerGrouping AS (SELECT 
   CustomerGroup = CASE 
        WHEN TotalOrderAmount >= 0 AND TotalOrderAmount <= 1000 THEN 'Low'
        WHEN TotalOrderAmount >= 1000 AND TotalOrderAmount <=5000 THEN 'Medium'
        WHEN TotalOrderAmount >=5000 AND TotalOrderAmount <=10000 THEN 'High'
        ELSE 'Very High'
    END
FROM TotalOrderAmount)

SELECT 
    CustomerGroup,
    TotalInGRoup = COUNT(*),
    PercentageInGroup = COUNT(*) * 1.00 / (SELECT COUNT(*) FROM CustomerGrouping)
FROM CustomerGrouping
GROUP BY CustomerGroup
ORDER BY TotalInGRoup DESC 
```
